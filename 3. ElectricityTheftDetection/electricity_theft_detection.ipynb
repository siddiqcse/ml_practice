```python
import pandas as pd
from sklearn.model_selection import train_test_split # for Data Separation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score, max_error, d2_tweedie_score
```

```python
# import data set by using pandas dataframe
df=pd.read_csv('elctricity_theft_detection.csv')
display(df.head())
display(df.isnull().sum())
print("Before SHAPE",df.shape)
```

```python
# remove CONS_NO column because this is not important for our prediction
df=df.drop(['CONS_NO'],axis=1)
display(df.head())
```

```python
print("Before SHAPE",df.shape)
columns1= df.columns.drop('FLAG')
# Remove Row when All row in columns are zero except FLAG Columns
df = df[(df[columns1] != 0).any(axis=1)]
print("After SHAPE",df.shape)
```

```python
# check is value NaN  if true then replace this Nan value by using mean techniques
for col in columns1:
    df[col] = df[col].fillna(df[col].mean())
```

```python
display(df.isnull().sum())
```

```python
# seperate training and testing data for train and test
y=df['FLAG'] # Store FLAG Data On Y
x=df.drop('FLAG',axis=1)
xTrain,xTest,yTrain,yTest=train_test_split(x,y,random_state=42,test_size=0.20)
```

```python
l = []
def metrics(model_name, y_pred_test_x,score):

    try:
        mean = mean_absolute_error(yTest, y_pred_test_x)
    except Exception as e:
        mean = 0

    try:
        r2 = r2_score(yTest, y_pred_test_x)
    except Exception as e:
        r2 = 0

    try:
        rmse = mean_squared_error(yTest, y_pred_test_x, squared=False)
    except Exception as e:
        rmse = 0


    try:
        explained_variance = explained_variance_score(yTest, y_pred_test_x, sample_weight=None, multioutput='uniform_average')
    except Exception as e:
        explained_variance = 0

    try:
        max_error_ = max_error(yTest, y_pred_test_x)
    except Exception as e:
        max_error_ = 0

    try:
        d2_tweedie_score_ =  d2_tweedie_score(yTest, y_pred_test_x)
    except Exception as e:
        d2_tweedie_score_ = 0

    print("Mean Absolute Error: ",mean )
    print("Mean Squared Error: ",rmse )
    print("R2 score: ",r2 )
    print("Explained Variance: ",explained_variance )
    print("max_error: ",max_error_ )
    print("d2_tweedie_score: ",d2_tweedie_score_ )
    print("Prediction Score  (PS):: ", score)

    l.append([model_name,score,  mean,rmse,  r2,explained_variance,max_error_ ,d2_tweedie_score_])
```

```python
from sklearn.ensemble import RandomForestRegressor
modelRFR = RandomForestRegressor(n_estimators = 100, random_state = 0)
modelRFR.fit(xTrain, yTrain)
metrics("Random Forest Regressor",modelRFR.predict(xTest),modelRFR.score(xTrain, yTrain))
print(modelRFR.score(xTest,yTest))
df1 = pd.DataFrame(l, columns=["Model Name", "Score", "mean", "rmse","r2","explained_variance","max_error","d2_tweedie_score"])
df1
```

```python
#Creation of DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
modelDTC = DecisionTreeClassifier()
modelDTC.fit(xTrain, yTrain)
metrics("DecisionTreeClassifier",modelDTC.predict(xTest),modelDTC.score(xTrain, yTrain))
print(modelDTC.score(xTest,yTest))
df1 = pd.DataFrame(l, columns=["Model Name", "Score", "mean", "rmse","r2","explained_variance","max_error","d2_tweedie_score"])
df1
```
