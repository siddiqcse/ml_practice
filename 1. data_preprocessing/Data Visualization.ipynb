```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import warnings as wr # for Handle Error 
import numpy as np
import matplotlib.pyplot as plt
# reading the database
data = pd.read_csv("train.csv")

# printing the top 10 rows
display(data.head(5))

cols = ['PassengerId','Name','Ticket','Cabin']
data = data.drop(cols,axis=1)

# printing the top 10 rows
display(data.head(5))

le=LabelEncoder()
wr.filterwarnings('ignore')
for col in data.columns:
    if data[col].dtype==np.number:
        continue
    data[col]=le.fit_transform(data[col])

# printing the top 10 rows
display(data.head(5))

data = data.dropna(axis=0)
```

```python
# Scatter Plot

# Scatter plot with day against tip
plt.scatter(data['Pclass'], data['Survived'],marker='o')
# Adding Title to the Plot
plt.title("Scatter Plot")
# Setting the X and Y labels
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.show()

plt.scatter(data['Survived'], data['Pclass'])
# Adding Title to the Plot
plt.title("Scatter Plot")
# Setting the X and Y labels
plt.xlabel('Survived')
plt.ylabel('Pclass')
plt.show()
```

```python
# Scatter plot with day against tip
plt.scatter(data['Survived'], data['Age'], c=data['Pclass'],
            s=data['SibSp'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.colorbar()
 
plt.show()
```

```python
# Scatter plot with day against tip

plt.plot(data['Survived'])
plt.plot(data['Age'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Survived')
plt.ylabel('Tip')
 
plt.show()
```

```python
# Bar chart with day against tip
plt.bar(data['Survived'], data['Age'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('Survived')
plt.ylabel('Age')
 
# Adding the legends
plt.show()
```

```python
# histogram of total_bills
plt.hist(data['Age'])
 
plt.title("Histogram")
 
# Adding the legends
plt.show()
```

# Bokeh
Bokeh is mainly famous for its interactive charts visualization. Bokeh renders its plots using HTML and JavaScript that uses modern web browsers for presenting elegant, concise construction of novel graphics with high-level interactivity.

```python
# importing the modules
from bokeh.plotting import figure, output_file, show
from bokeh.palettes import magma
```

```python
# instantiating the figure object
graph = figure(title = "Bokeh Scatter Graph")
color = magma(256)
 
# plotting the graph
graph.scatter(data['Survived'], data['Age'], color=color)
 
# displaying the model
show(graph)
```

```python
# Line Chart
# instantiating the figure object
graph = figure(title = "Bokeh Bar Chart")
 
# Count of each unique value of
# tip column
df = data['Pclass'].value_counts()
# plotting the graph
graph.line(df, data['Pclass'])
# displaying the model
show(graph)
```

```python
# Bar Chart
graph = figure(title = "Bokeh Bar Chart")
 
# Count of each unique value of
# tip column
df = data['Pclass'].value_counts()
# plotting the graph
graph.vbar(data['Age'], top=data['Survived'])
# displaying the model
show(graph)
```
