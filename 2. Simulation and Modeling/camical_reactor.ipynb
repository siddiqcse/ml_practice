```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(columns=["Time", "A(I)", "B(I)","C(I)"])

a=100.00
b=50.00
c=0.00
k1=0.008
k2=0.002
time=5
delT=0.1
N=10
countTime=0.00
df.loc[len(df)] = [countTime, a, b,c]


for x in range(0, N):
  newA=a+((k2*c)-(k1*a*b))*delT
  newB=b+((k2*c)-(k1*a*b))*delT
  newC=c+(2*((k1*a*b)-(k2*c)))*delT
  countTime=round(countTime+delT,2)
  a=round(newA,2)
  b=round(newB,2)
  c=round(newC,2)
  df.loc[len(df)] = [countTime, a, b,c]



display(df)

# # histogram of total_bills
plt.hist(df['C(I)'])
 
plt.title("Histogram")
 
# Adding the legends
plt.show()
sns.histplot(x='C(I)', data=df, kde=True)
 

```
