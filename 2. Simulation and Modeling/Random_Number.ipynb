```python
# Task 2
# Linear Congruential Generator (LCG) for PRANGs
import matplotlib.pyplot as plt
import random   # to generate arbitary values of a , c and m
seed=int(input("Enter a seed: "))
RN=int(input("Enter number of RNs to be generated: "))
number=seed
m=random.randint(1,100)
a=random.randint(1,100)
c=random.randint(1,100)
x=[]
for i in range(RN):
    if ((a*number)%m)!=0:
        number=(a*number+c)%m
        x.append(number)
print(x)
plt.scatter(range(RN),x)
plt.grid()
plt.show()

# Xn+1= (aXn+c) mod m
# (aXn+c) mod m should not be zero. It also means that (aXn+c) and m should be relatively prime
```

```python
# Kolmogorov Smirnov Test
seq = [[float(x)] for x in input("Enter the random sequece: ").split(',') if 0<=float(x)<1]
alpha = float(input("Enter the critical value: "))
seq.sort()                                      #seq[i][0]: R(i)
N = len(seq)

for i in range(N):
    seq[i].append((i+1)/N)                      #seq[i][1]: i/N
    seq[i].append(seq[i][1]-seq[i][0])          #seq[i][2]: i/N - R(i)
    seq[i].append(seq[i][0]-i/N)                #seq[i][3]: R(i) - (i-1)/N
    print(seq[i])

Dplus = max([s[2] for s in seq if s[2]>=0])
Dminus = max([s[3] for s in seq if s[3]>=0])
D = max(Dplus, Dminus)

if D < alpha:
    print("\nThe random sequence is uniformly distributed")
else:
    print("\nThe random sequence is not uniformly distributed")
```
