```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
vf=20
xf=0
yf=50

time=0
xb=[80,90,99,108,116,125,133,141,151,160,196,179,180]
yb=[0,-2,-5,-9,-15,-18,-23,-29,-28,-25,-21,-20,-17]
list=[0]
list1=[50]

list2=[80]
list3=[0]
flag=0
while flag<=13:
  dis=math.sqrt((xf-xb[time])**2+(yf-yb[time])**2)


  if dis<=10:
    print("Bomber plan shotdown",time)
    break
  elif time>11:
     print("Bomber plan escape",time)
     break
  else:
     xf=xf+vf*(xb[time]-xf)/dis
     list.insert(time+1,xf)

     yf=yf+vf*(yb[time]-yf)/dis
     list1.insert(time+1,yf)


     list2.insert(time+1,xb[time])
     list3.insert(time+1,yb[time])
     time+=1
  flag=flag+1


plt.grid()
plt.plot(list,list1,label='Fighter')
plt.plot(list2,list3,label='Bomber')

```
